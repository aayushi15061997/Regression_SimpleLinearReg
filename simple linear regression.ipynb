```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

dataset=pd.read_csv('/Users/Aayushi/Desktop/ML/Regression/Simple_Linear_Regression/Salary_Data.csv')
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,1].values
dataset

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=1/3,random_state=0)
#no feature scaling is req. as the lib takes care of that for us

#fitting slr to the training set
#import this lib for slr
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
#predicting the test set results
#to do this we create a vector where we store the predicted result
y_pred = regressor.predict(x_test)

y_pred,y_test

```

```python
#to display in table form
from tabulate import tabulate

m = np.array([y_pred, y_test])
headers = ["y_pred", "y_test"]

# tabulate data
table = tabulate(m, headers, tablefmt="simple")
print(table)

```

```python
#visualising the training set results
#using matplotlib we plot the points here in red color
plt.scatter(x_train,y_train,color='red')
#we plot the regression line here
#'y' coordinates of this plot func will be the prediction of the x_train

plt.plot(x_train,regressor.predict(x_train),color='blue')
plt.title('salary vs experience(training set)')
plt.xlabel('years of experience')
plt.ylabel('salary')
plt.show()
```

```python
#visualising the test set results
#our regressor is already trained on training sets
#so whether we keep x_train or replace it by x_test in the second line
#it will give us the same unique equation
plt.scatter(x_test,y_test,color='red')
plt.plot(x_train,regressor.predict(x_train),color='blue')
plt.title('salary vs experience(test set)')
plt.xlabel('years of experience')
plt.ylabel('salary')
plt.show()
```
